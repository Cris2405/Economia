```python
import pandas as pd

# Lee el archivo Excel y carga solamente las columnas especificadas
df = pd.read_excel('/content/Data_Bonos.xlsx', usecols=['CUPON', 'NOMBRE EMISOR', 'countrydomicile', 'INDUSTRIA', 'FECHA EMISION', 'YIELD', 'TIPO CUPON', 'PRINCIPAL'])

# Filtrar los datos segÃºn las condiciones especificadas
datos_filtrados = df[(df['INDUSTRIA'] == 'Tipo 1') & ((df['CUPON'] > 7) | (df['YIELD'] > 8))]
df
```

```python
import pandas as pd

# Creamos una lista con las compaÃ±Ã­as en el orden deseado
orden_compa = ["INTERNATIONAL BUSINESS MACHS CORP", "TIME WARNER COS INC", "WALMART STORES INC"]

# Filtramos los datos para incluir solamente las compaÃ±Ã­as en la lista orden_compaÃ±Ã­as
df_filtrado = df[df['NOMBRE EMISOR'].isin(orden_compa)]

# Ordenamos los datos segÃºn el orden de la lista orden_compaÃ±Ã­as
df_ordenado = df_filtrado.sort_values("NOMBRE EMISOR", key=lambda x: pd.Series(x).map(lambda y: orden_compa.index(y)))

df_ordenado
```

```python
import pandas as pd

# Ordenamos los datos segÃºn la columna "FECHA EMISION" en orden descendente
df_ordenado = df.sort_values("FECHA EMISION", ascending=False)

# Imprimimos los datos ordenados
print(df_ordenado)
```

```python
def formato(valor):
    rojo = 'background-color: red'
    default = ''
    return pd([rojo if i > 8 else default for i in valor])

# Aplicar formato a la columna "YIELD"
df.style.apply({'YIELD': formato}, axis=0)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Precios_Acciones.xlsx')
df
```

```python
plt.plot(df['Date'], df['IBM'])
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.title('Precio de las acciones de IBM')
plt.show()
```

```python
plt.plot(df['Date'], df['IBM'], label='IBM')
plt.plot(df['Date'], df['Google'], label='Google')
plt.ylabel('Precio de cierre')
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()

# Graficar los precios de las acciones de IBM en el eje Y principal
ax1.plot(df['IBM'], label='IBM', color='b')
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Precio de la acciÃ³n (IBM)', color='b')
ax1.tick_params(axis='y', colors='b')

# Crear un segundo eje Y para los precios de las acciones de Google
ax2 = ax1.twinx()
ax2.plot(df['Google'], label='Google', color='r')
ax2.set_ylabel('Precio de la acciÃ³n (Google)', color='r')
ax2.tick_params(axis='y', colors='r')

# AÃ±adir la leyenda y el tÃ­tulo
fig.legend()
plt.title('Precios de las acciones de IBM y Google')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

returns = df.pct_change()

# Crear el grÃ¡fico de dispersiÃ³n
plt.scatter(returns["Google"], returns["Berkshire Hathaway"])
plt.xlabel("Rendimientos diarios de Google")
plt.ylabel("Rendimientos diarios de Berkshire Hathaway")
plt.title("GrÃ¡fico de dispersiÃ³n entre los rendimientos de las acciones de Google y Berkshire Hathaway")
plt.show()
```

```python
# Calcular las rentabilidades medias y mÃ¡ximas
mean_returns = df.mean()
max_returns = df.max()

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Graficar la rentabilidad media en un grÃ¡fico de columnas
ax1.bar(mean_returns.index, mean_returns.values, alpha=0.5, color='b')
ax1.set_xlabel('Acciones')
ax1.set_ylabel('Rentabilidad media', color='b')
ax1.tick_params('y', colors='b')

# Crear un segundo eje Y para la rentabilidad mÃ¡xima
ax2 = ax1.twinx()
ax2.plot(max_returns.index, max_returns.values, color='r')
ax2.set_ylabel('Rentabilidad mÃ¡xima', color='r')
ax2.tick_params('y', colors='r')

# Ajustar los lÃ­mites del eje Y para el grÃ¡fico de columnas
ax1.set_ylim([0, 0.03])

# AÃ±adir el tÃ­tulo y mostrar la figura
plt.title('Rentabilidad media y mÃ¡xima de las acciones')
plt.show()
```
