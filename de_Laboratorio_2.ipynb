```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Precios_Acciones.xlsx')
df
```

```python
plt.plot(df['Date'], df['IBM'])
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.title('Precio de las acciones de IBM')
plt.show()
```

```python
plt.plot(df['Date'], df['IBM'], label='IBM')
plt.plot(df['Date'], df['Google'], label='Google')
plt.ylabel('Precio de cierre')
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()

# Graficar los precios de las acciones de IBM en el eje Y principal
ax1.plot(df['IBM'], label='IBM', color='b')
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Precio de la acción (IBM)', color='b')
ax1.tick_params(axis='y', colors='b')

# Crear un segundo eje Y para los precios de las acciones de Google
ax2 = ax1.twinx()
ax2.plot(df['Google'], label='Google', color='r')
ax2.set_ylabel('Precio de la acción (Google)', color='r')
ax2.tick_params(axis='y', colors='r')

# Añadir la leyenda y el título
fig.legend()
plt.title('Precios de las acciones de IBM y Google')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

returns = df.pct_change()

# Crear el gráfico de dispersión
plt.scatter(returns["Google"], returns["Berkshire Hathaway"])
plt.xlabel("Rendimientos diarios de Google")
plt.ylabel("Rendimientos diarios de Berkshire Hathaway")
plt.title("Gráfico de dispersión entre los rendimientos de las acciones de Google y Berkshire Hathaway")
plt.show()
```

```python
# Calcular las rentabilidades medias y máximas
mean_returns = df.mean()
max_returns = df.max()

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Graficar la rentabilidad media en un gráfico de columnas
ax1.bar(mean_returns.index, mean_returns.values, alpha=0.5, color='b')
ax1.set_xlabel('Acciones')
ax1.set_ylabel('Rentabilidad media', color='b')
ax1.tick_params('y', colors='b')

# Crear un segundo eje Y para la rentabilidad máxima
ax2 = ax1.twinx()
ax2.plot(max_returns.index, max_returns.values, color='r')
ax2.set_ylabel('Rentabilidad máxima', color='r')
ax2.tick_params('y', colors='r')

# Ajustar los límites del eje Y para el gráfico de columnas
ax1.set_ylim([0, 0.03])

# Añadir el título y mostrar la figura
plt.title('Rentabilidad media y máxima de las acciones')
plt.show()
```
